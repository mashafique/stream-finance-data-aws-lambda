```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('results.csv')
```

```python
data
```

# Highest Stock Price at the First Trading Hour

```python
data_at_9 = data[data['Hour'] == 9]
data_at_9
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='Company', y='Hourly_High', data=data_at_9)
plt.xlabel("Company", fontsize=20)
plt.ylabel("Stock Price", fontsize=20)
plt.title("Highest Stock Price at the First Trading Hour", fontsize=20)
```

# Highest Hourly Stock Price Trend

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Hour', y='Hourly_High', data=data, hue='Company')
plt.xlabel("Company", fontsize=20)
plt.ylabel("Stock Price", fontsize=20)
plt.title("Highest Stock Price Trend", fontsize=20)
plt.legend(loc='upper right')
```
